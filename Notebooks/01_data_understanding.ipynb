## Feature Categorization

To better understand the dataset, we categorize the features into different groups based on the type of information they provide.


```python
feature = [col for col in df.columns if col != 'Churn Value']
numeric = df[feature].select_dtypes(include=['int64','float64']).columns.tolist()
categorical = df[feature].select_dtypes(include=['object']).columns.tolist()
target = 'Churn Value'

print(f'Feature Columns: {feature}')
print(f'Numeric Columns: {numeric}')
print(f'Categorical Columns: {categorical}')
print(f'Target Variable: {target}')
```
