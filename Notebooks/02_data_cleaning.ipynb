```python
df.columns=df.columns.str.lower().str.replace(" ","_").str.strip()
df.columns
```

#Column Selection
###Here i willl categorize all columns in 3 parts.
target - which have the target variable\
real - it has all the real columns which should be used as it contains all the necessary informations\
drop - all the unnecessary and noisy columns which can confuse the model and doesn't makek any changes.\

###drop columns consist of
identifier - customerid
constans - count ,country\
location - zip_codes , etc\
lekage - churn_score ,cltv (as can be identified after prediction only)

###we will keep
city because it will give us the location of the users and we can use them


```python
target=df['churn_value']
real=df[['city','gender', 'senior_citizen',
       'partner', 'dependents', 'tenure_months', 'phone_service',
       'multiple_lines', 'internet_service', 'online_security',
       'online_backup', 'device_protection', 'tech_support', 'streaming_tv',
       'streaming_movies', 'contract', 'paperless_billing', 'payment_method','monthly_charges','total_charges','churn_reason']]
drop=df[['customerid','count', 'country', 'state', 'zip_code',
       'lat_long', 'latitude', 'longitude', 'churn_label', 'churn_value',
       'churn_score','cltv']]
```

```python
data=pd.concat([real,target],axis=1)
data.head()
```

#Removing the duplicates/errors
Sometimes data have duplicate rows which should be removed as can disturb the skewness of data , moreover they can have an empty space which can act as a missing value but we can't fix it with traditional fillna methods thatswhy it has to be replaced with np.nan...

```python
data=data.replace(['',' '],np.nan)
data.drop_duplicates(inplace=True)
```

#Dealing with Missing Values

```python
data.info()
```

```python
data.isna().sum()
```

Here as we can see we have missing values in two columns but we will only deal with **total_charges** because **churn_reason** has only valid reasons for the chustomers who have churned and remaining customers have nan values.


```python
print(f'total rows in churn_reason: {data.shape[0]}')
print(f'total churned chustomers: {data[data['churn_value']==1].shape[0]}')
print(f'total non churned chustomers: {data[data['churn_value']==0].shape[0]}')
print(f'total missing values in total_charges: {data["churn_reason"].isna().sum()}')
print('hence valid missing values')
```

###Replacing The MV in Total Charges

```python
data[data['total_charges'].isna()].head()
```

```python
data['total_charges'].fillna(data['total_charges'].median(), inplace=True)
data['total_charges'].isna().sum()
```
